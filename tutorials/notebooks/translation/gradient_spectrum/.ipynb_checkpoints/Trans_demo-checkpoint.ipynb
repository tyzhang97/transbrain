```python
import pandas as pd
import numpy  as np
from nilearn import image,plotting
from scipy import stats
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
```

```python
import TransBrain as TB
```

## Gradient

```python
Mouse_atlas = TB.atlas.fetch_mouse_atlas(region_type='all')
```

```python
mouse_gradient_array = np.zeros((3,68))
for i in range(3):
    gradient_img = image.load_img('Gradient_{}_in_CCFv3.nii.gz'.format(i+1))
    mouse_gradient_phenotype = TB.base.get_region_phenotypes(gradient_img,species='Mouse')
    mouse_gradient_array[i] = mouse_gradient_phenotype.values.flatten()
```

```python
mouse_gradient = pd.DataFrame(mouse_gradient_array,index=['Gradient1','Gradient2','Gradient3'],columns=Mouse_atlas['info_table']['Anatomical Name'].values)
```

```python
mouse_gradient = mouse_gradient.T
```

### translating mouse gradient to human

```python
Transformer = TB.trans.SpeciesTrans()
```

```python
gradient_trans_in_human = Transformer.mouse_to_human(mouse_gradient,region_type='cortex')
```

### visulization

```python
Human_atlas = TB.atlas.fetch_human_atlas(region_type='all')
```

### mouse gradient

```python
m_g_1 = image.load_img('Gradient_1_in_CCFv3.nii.gz')
m_g_2 = image.load_img('Gradient_2_in_CCFv3.nii.gz')
m_g_3 = image.load_img('Gradient_3_in_CCFv3.nii.gz')
```

```python
Mouse_template = image.load_img('../../../Atlas/P56_Atlas.nii.gz')
```

```python
plotting.plot_stat_map(m_g_1,bg_img=Mouse_template,draw_cross=False,display_mode='y',
                       annotate=True,cut_coords=range(-4,3,1),cmap='coolwarm',symmetric_cbar=True)
```

```python
plotting.plot_stat_map(m_g_2,bg_img=Mouse_template,draw_cross=False,display_mode='y',
                       annotate=True,cut_coords=range(-4,3,1),cmap='coolwarm',symmetric_cbar=True)
```

```python
plotting.plot_stat_map(m_g_3,bg_img=Mouse_template,draw_cross=False,display_mode='y',
                       annotate=True,cut_coords=range(-4,3,1),cmap='coolwarm',symmetric_cbar=True)
```

```python
graph_vis_map_l = []

for col in ['Gradient1','Gradient2','Gradient3']:

    graph_map = stats.zscore(gradient_trans_in_human[col].values)
    
    graph_zero_data = np.zeros_like(Human_atlas['atlas_data'])
    
    for i,ele in enumerate(graph_map):
        
        graph_zero_data[Human_atlas['atlas_data']==i+1] = ele
        
    graph_vis_map = image.new_img_like(Human_atlas['atlas'],graph_zero_data)
    graph_vis_map_l.append(graph_vis_map)
```

```python
plotting.plot_stat_map(graph_vis_map_l[0],title='Mouse Gradient1 Mapping',cmap='coolwarm',cut_coords=(-27,-10,19))
```

```python
plotting.plot_stat_map(graph_vis_map_l[1],title='Mouse Gradient2 Mapping',cmap='coolwarm',cut_coords=(-27,-10,19))
```

```python
plotting.plot_stat_map(graph_vis_map_l[2],title='Mouse Gradient3 Mapping',cmap='coolwarm',cut_coords=(-27,-10,19))
```

### correlation

```python
human_matrix_gradient = []
for g in [h_g_1,h_g_2,h_g_3]:
    human_matrix_gradient.append(TB.base.get_region_phenotypes(g)['Phenotype'].values[:105])
```

```python
human_matrix_gradient = np.asarray(human_matrix_gradient)
```

```python
similarity_matrix_gradient = np.zeros((3,3))
for i,mouse_v in enumerate(gradient_trans_in_human.T.values):
    for j,human_v in enumerate(human_matrix_gradient):
        r = stats.pearsonr(mouse_v,human_v)[0]
        similarity_matrix_gradient[i,j] = r
```

```python
similarity_matrix_gradient = pd.DataFrame(similarity_matrix_gradient,
                                          index=['Mouse gradient{}'.format(i+1) for i in range(3)],
                                         columns=['Human gradient{}'.format(i+1) for i in range(3)])
```

```python
similarity_matrix_gradient
```

```python
sns.heatmap(similarity_matrix_gradient,cmap='RdBu_r',annot=True)
```
